Genera dos exceles para copiar los valores de los diametros maximos y las records donde se localizan dichos diametros

```python
import os
#import pickle
import json
import shutil

import pandas as pd
```

```python
data_excel_path = 'C:/PROJECTS/statistics/data/excels/'
save_excel_path = 'C:/PROJECTS/statistics/data/save_pd.xlsx'
path_results = 'C:/PROJECTS/ENDOCARDITIS/macro_emboendoIA/data/results/results_'
path_summaries = 'C:/PROJECTS/ENDOCARDITIS/macro_emboendoIA/data/summaries/'
path_results_test_risk = 'C:/PROJECTS/statistics/data/peakles_measured_AI/'
path_summaries_save = 'C:/PROJECTS/statistics/data/summaries/'
```

```python
excel_file_in_name = 'data_29_01_26.xlsx'
splits_n = 10
splits = [str(split) for split in range(1,splits_n +1)]
```

# CLEAN DIRS

```python
for pth in [path_results_test_risk,path_summaries_save]:
    for file in os.listdir(pth):
        os.remove(pth+file)
```

# MOVE FILES

```python
for split in splits:
    origin = path_summaries + f'results_split_{split}.json'
    destination = path_summaries_save + f'results_split_{split}.json'
    shutil.copy(origin, destination)
```

```python
split_patients = {}
for split in splits:
    files = os.listdir(path_results + split)
    patients = []
    for file in files:
        patients.append(file.split('.')[0])
        origin = path_results + split + '/' + file
        destination = path_results_test_risk + file
        shutil.copy(origin, destination)
    split_patients[split] = patients
    print(f'Patients: {len(patients)} in split {split}')
```

```python
patients_diameters = {}
patients_records =  {}
for file in os.listdir(path_summaries_save):
    with open(path_summaries_save + file, 'rb') as f:
        data = json.load(f)

    for results in data.values():
        for patient,result in results.items():
            patients_diameters[patient] = result['diam']
            patients_records[patient] = result['measured_record']
```

```python
for pat,diam in patients_diameters.items():
    print(f'{pat}: {patients_records[pat]} : {diam} mm')
```

# GENERATE EXCEL

```python
df_monitorizacion = pd.read_excel(data_excel_path + excel_file_in_name, decimal = ',')
```

```python
df_monitorizacion = df_monitorizacion[['record_id', 'PatientID','Final','IA','IA_RECORD']]

df_monitorizacion['PatientID']= df_monitorizacion['PatientID'].astype(str)
df_monitorizacion['record_id']= df_monitorizacion['record_id'].astype(str)

df_monitorizacion_anom = df_monitorizacion.copy()
```

### diameters and records

```python
df_monitorizacion['IA'] = df_monitorizacion['PatientID'].map(patients_diameters)
df_monitorizacion['IA_RECORD'] = df_monitorizacion['PatientID'].map(patients_records)
```
